#### **Desafio**

Suspeita de 26% de desistência de clientes com uma empresa. Como solucionar isso, uma vez que isso indica perda de valores?



```python
import pandas as pd 
```

```python
#Importação de dados e visualização da base de dados
frame = pd.read_csv(r'data\telecom_users.csv')
display(frame)
```

### **Tratamento de dados**

a) Informações inúteis devem ser retiradas. 
        
        colunas com valores vazios; 
        informações que não acrescentam em nada;

b) Valores reconhecidas de maneiras erradas. 

        Verificar se como o python esta reconhecendo os tipos de dado em cada coluna. 

c) Tratar os NaN. 

```python
# a) Deleção de linhas e de colunas de uma tabela 
# axis = 0 --> linha
# axis = 1 --> coluna 

frame = frame.drop(frame.columns[0], axis=1)
```

```python
# b) Non-Null --> informa a quantidade de linhas preenchidas em cada coluna 
# Dtype: object similar a string

frame.info()
```

```python
#Tranformar valores 'object' para númericos

frame['TotalGasto'] = pd.to_numeric(frame['TotalGasto'], errors = 'coerce')

#Parâmetro coerce é extremamente importante, uma vez que, se tiver um erro de digitação (entradas que não são números), 
#o código acrescenta 'NaN'.

display(frame.info())
```

```python
#c) Tratamento de valores vazios

#all = Deletar somente linhas/colunas cujos valores são vazios 
#any = Deletar linhas/colunas com pelo menos 1 valor vazio 

frame.dropna(how='all',axis=1)
frame.dropna(how='any', axis=0)

display(frame)
```

```python
#Quantificar quantas pessoas desistiram 

display(frame['Churn'].value_counts())
display(frame['Churn'].value_counts(normalize=True))
```
